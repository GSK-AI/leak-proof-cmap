# Read in split data and summarise


```python
from pathlib import Path
import pandas as pd

working_dir=Path("working_dir")
split_data=(working_dir/Path("split_data")).resolve()
metric_performance_df = pd.read_csv(split_data / "split_info.tsv", sep="\t")
metric_performance_df["split"] = metric_performance_df["cli"].astype(str) + "," + metric_performance_df["moai"].astype(str)
split_counts_summary_df = pd.DataFrame()
split_counts_summary_df.index.name = "split"
for cli in range(1, 6):
    for moai in range(1, 6):
        row_train=metric_performance_df.query(f"cli=={cli} and moai=={moai} and fold_type=='train' and fold_i==1.0")
        row_val=metric_performance_df.query(f"cli=={cli} and moai=={moai} and fold_type=='val' and fold_i==1.0")
        row_test=metric_performance_df.query(f"cli=={cli} and moai=={moai} and fold_type=='test'")
        split_counts_summary_df.loc[
            f"{cli},{moai}",
            [
                "Training and validation profile count",
                "Training and validation unique MOA count",
                "Training and validation unique compound count",
                "Test profile count",
                "Test unique MOA count",
                "Test unique compound count",
            ],
        ] = (
            row_train['n_records'].item()+row_val['n_records'].item(),
            row_train['unique_moas'].item()+row_val['unique_moas'].item(),
            row_train['unique_compounds'].item()+row_val['unique_compounds'].item(),
            row_test['n_records'].item(),
            row_test['unique_moas'].item(),
            row_test['unique_compounds'].item(),
        )
display(split_counts_summary_df.astype(int))
split_counts_summary_df.to_csv(split_data/"split_info_unique_counts_within_splits.csv")
```
